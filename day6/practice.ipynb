This notebook covers Python list concepts, methods, slicing, and practical applications.


```python
# Creating a basic Python list
my_list = [1, 2, 3, "text", True]
print(my_list)
```

## List Properties
- Ordered collection
- Mutable (can be changed)
- Allows mixed data types


```python
# Indexing and negative indexing
print("First element:", my_list[0])
print("Last element:", my_list[-1])
```

## Accessing List Items Using Slicing

```python
# List slicing examples
numbers = [10, 20, 30, 40, 50, 60]
print(numbers[1:4])
print(numbers[::2])
```

## Adding Items to a List
- append()
- insert()
- extend()


```python
students = ["Ali", "Sara", "Ahmed"]

students.append("Ayesha")
students.insert(1, "Bilal")
students.extend(["Hassan", "Zain"])

print(students)
```

## Removing Items from a List
- remove()
- pop()
- clear()

```python
students.remove("Sara")
removed_student = students.pop()
print("Updated list:", students)
print("Removed student:", removed_student)
```
