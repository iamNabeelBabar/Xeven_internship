```python
#Day 8   ----->Students Grade Manager

# -----------------------------
# Student Grade Manager
# -----------------------------

students = ["Ali", "Sara", "John", "Emma", "David"]
grades = [85, 92, 78, 88, 95]


def add_student(name, grade):
    students.append(name)
    grades.append(grade)


def remove_student(name):
    if name in students:
        index = students.index(name)
        students.pop(index)
        grades.pop(index)


def update_grade(name, new_grade):
    if name in students:
        index = students.index(name)
        grades[index] = new_grade


def get_average():
    return sum(grades) / len(grades)


def highest_lowest():
    max_grade = max(grades)
    min_grade = min(grades)
    return (
        students[grades.index(max_grade)], max_grade,
        students[grades.index(min_grade)], min_grade
    )


def sort_by_grade_desc():
    combined = list(zip(students, grades))
    combined.sort(key=lambda x: x[1], reverse=True)
    return combined


# Example Execution
add_student("Hassan", 90)
update_grade("Ali", 89)

avg = get_average()
top_sorted = sort_by_grade_desc()

print("Average:", avg)
print("Top 3 Performers:", top_sorted[:3])

above_avg = [students[i] for i in range(len(grades)) if grades[i] > avg]
below_avg = [students[i] for i in range(len(grades)) if grades[i] <= avg]

print("Above Average:", above_avg)
print("Below Average:", below_avg)

```

# **Shopping Cart Management**

```python
# Shopping Cart System

cart = []


def add_item(name, price, quantity):
    item = {
        "name": name,
        "price": price,
        "quantity": quantity
    }
    cart.append(item)


def remove_item(name):
    for item in cart:
        if item["name"] == name:
            cart.remove(item)
            break


def update_quantity(name, quantity):
    for item in cart:
        if item["name"] == name:
            item["quantity"] = quantity
            break


def calculate_total():
    total = 0
    for item in cart:
        total += item["price"] * item["quantity"]

    if total > 100:
        total = total * 0.9   # apply 10% discount

    return total


def display_receipt():
    print("\nReceipt")

    for item in cart:
        subtotal = item["price"] * item["quantity"]
        print(f"{item['name']}  Qty: {item['quantity']}  "
              f"Price: {item['price']}  Subtotal: {subtotal}")

    print("Total:", calculate_total())


# Example usage
add_item("Laptop", 80, 1)
add_item("Mouse", 15, 2)
add_item("Keyboard", 30, 1)
add_item("Headphones", 25, 1)

display_receipt()

print("\nRecently added items:", cart[-3:])
```
