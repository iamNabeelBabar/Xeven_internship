```python
# 3. Returning multiple values from a function
def min_max(lst):
    return min(lst), max(lst)

low, high = min_max([4, 7, 1, 9])
print("Min:", low, "Max:", high)
```

**Set practice**

```python
# 6. Creating sets
visitors_day1 = {"192.168.0.1", "192.168.0.2", "192.168.0.3"}
visitors_day2 = {"192.168.0.2", "192.168.0.3", "192.168.0.4"}
print(visitors_day1)
```

```python
# 7. Union of sets
all_visitors = visitors_day1.union(visitors_day2)
print("All unique visitors:", all_visitors)
```

```python
# 8. Intersection of sets
common_visitors = visitors_day1.intersection(visitors_day2)
print("Common visitors:", common_visitors)
```

```python
# 9. Difference of sets
unique_day1 = visitors_day1.difference(visitors_day2)
print("Unique to day1:", unique_day1)
```

```python
# 10. Symmetric difference
unique_visitors = visitors_day1.symmetric_difference(visitors_day2)
print("Visitors only on one day:", unique_visitors)
```

**Combined tuples and sets**

```python
# 11. List of city tuples and extracting names into a set
cities = [("NY", 40.7, -74.0), ("LA", 34.0, -118.2), ("Chicago", 41.8, -87.6)]
city_names = {city[0] for city in cities}
print("City names:", city_names)
```

```python
# 12. Tracking unique emails
emails = ["abc@gmail.com", "xyz@yahoo.com", "abc@gmail.com"]
unique_emails = set(emails)
print("Unique emails:", unique_emails)
```

```python
import math

def distance(coord1, coord2):
    lat1, lon1 = coord1[1], coord1[2]
    lat2, lon2 = coord2[1], coord2[2]
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dlon/2)**2
    return 6371 * 2 * math.asin(math.sqrt(a))

my_loc = ("MyPlace", 41.0, -87.0)
nearest = min(cities, key=lambda c: distance(my_loc, c))
print("Nearest city:", nearest[0])
```
